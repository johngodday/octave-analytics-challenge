### Question 1 The airline that had most number of travels

```python
data.groupby('OP_UNIQUE_CARRIER')['FLIGHTS'].sum().reset_index().sort_values(by = "FLIGHTS")
```

The airline that had most number of travels is WN with a total of 114442

### Question 2 The total number of flights between in 2019

```python
data.groupby('OP_UNIQUE_CARRIER')['FLIGHTS'].sum().sum()
```

### Question 3 What is the Total AirTime by all Airlines in 2019 (answer in hours)

```python
air_time = data['AIR_TIME']
air_time_sum = data['AIR_TIME'].sum()
air_time_sum
```

```python
hours = air_time_sum / 60

print('Total AirTime by all Airlines in December 2019 is {} hours'.format(hours))
```

### What airline covered the most distance in December 2019? (5 Points) Mention the Airline Name (not code) in CAPITAL LETTERS
